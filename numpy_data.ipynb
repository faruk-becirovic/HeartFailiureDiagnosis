```python
from keras.models import Sequential
from keras.layers import Dense, Activation

model  = Sequential()
model.add(Dense(32, activation='relu', input_shape=(19,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(48, activation='relu'))
model.add(Dense(48, activation='relu'))
model.add(Dense(48, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
[print(i.shape, i.dtype) for i in model.inputs]
[print(o.shape, o.dtype) for o in model.outputs]
[print(l.name, l.input_shape, l.dtype) for l in model.layers]
```

```python
import tensorflow as tf
data_set_train = tf.convert_to_tensor(data_set_train, dtype=tf.float32)
data_set_train_results = tf.convert_to_tensor(data_set_train_results, dtype=tf.float32)
```

```python
model.fit(data_set_train, data_set_train_results, epochs=500, batch_size=16)
```
